```python
import os
import pandas as pd
import PyPDF2


# PDF
def read_pdf_files(path):
    files = [f for f in os.listdir(path) if f.endswith('.pdf')]
    file_texts = []
    
    for file in files:
        pdf_file_obj = open(os.path.join(path, file), 'rb')
        pdf_reader = PyPDF2.PdfReader(pdf_file_obj)
        num_pages = len(pdf_reader.pages)
        text = ''
        for page in range(num_pages):
            page_obj = pdf_reader.pages[page] 
            text += page_obj.extract_text()
        file_texts.append((file,text))
        pdf_file_obj.close()
        
    return file_texts

def split_into_chunks(text, chunk_size):
    words = text.split()
    chunks = [' '.join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]

    print(chunks)

    # Do overlapped chunks
    chunks_overlapped = []
    window = 5  # number of segments to combine
    stride = 2  # number of segments to 'stride' over, used to create overlap
    for i in (range(0, len(chunks), stride)):
        i_end = min(len(chunks)-1, i+window)
        text = ' '.join(_ for _ in chunks[i:i_end])
        chunks_overlapped.append(text)
    return chunks_overlapped


def join_all_chunks(file_texts, chunk_size):
    chunks_list = []
    for file, text in file_texts:
        chunks = split_into_chunks(text, chunk_size)
        for chunk in chunks:
            chunks_list.append((file, chunk))

    return chunks_list

def convert_chunks_to_df(file_texts, chunk_size):
    chunks_list= join_all_chunks(file_texts, chunk_size)
    df = pd.DataFrame(chunks_list, columns=['file_name', 'text'])
    return df

file_texts = read_pdf_files(".././documents/pdf/")
df1 = convert_chunks_to_df(file_texts, 200)

df1
```

```python
import os
import pandas as pd

def read_and_transform_csv(path):
    files = [f for f in os.listdir(path) if f.endswith('.csv')]
    data = []

    for file in files:
        df = pd.read_csv(os.path.join(path, file))
        for _, row in df.iterrows():
            text = '; '.join([f'{col}:{val}' for col, val in row.to_dict().items()])
            data.append((file, text))

    return data



data = read_and_transform_csv(".././documents/csv/")

df2 = pd.DataFrame(data, columns=['file_name', 'text'])

df2.to_csv('data.csv', index=False)
df2
```

```python
import os
from openai.embeddings_utils import get_embedding, cosine_similarity
import tiktoken

# df = pd.concat([df1,df2])


df = df2

# embedding model parameters
embedding_model = os.getenv("DEPLOYMENT_EMBEDDED_NAME")
embedding_encoding = "cl100k_base"  # this the encoding for text-embedding-ada-002
max_tokens = 8000  # the maximum for text-embedding-ada-002 is 8191

encoding = tiktoken.get_encoding(embedding_encoding)

# remove any row with empty text
df = df[df.text.ne('')]

df["n_tokens"] = df.text.apply(lambda x: len(encoding.encode(x)))

df["embedding"] = df.text.apply(lambda x: get_embedding(x, engine=embedding_model))

df
```

```python
def prepare_prompt(prompt, results):
  tokens_limit = 4096 # Limit for gpt-3.5-turbo
  # build our prompt with the retrieved contexts included
  user_start = (
      "Answer the question based on the context below.\n\n"+
      "Context:\n"
  )

  user_end = (
      f"\n\nQuestion: {prompt}\nAnswer:"
  )

  count_of_tokens_consumed = len(encoding.encode("\"role\":\"system\"" + ", \"content\" :\"" + system
                                            + user_start + "\n\n---\n\n" + user_end))

  count_of_tokens_for_context = tokens_limit - count_of_tokens_consumed

  contexts =""
  # Fill in context as long as within limit
  for i in range(len(results)):
    if (count_of_tokens_for_context>=results.n_tokens.iloc[i]):
        contexts += results.text.iloc[i] + "\n"
        count_of_tokens_for_context -=1
        count_of_tokens_for_context -= results.n_tokens.iloc[i]

  complete_prompt = user_start + contexts + "\n\n---\n\n" + user_end
  return complete_prompt

def answer(messages):
  response = openai.ChatCompletion.create(
              engine="35-turbo-16",
              messages=messages,
              temperature=0.3
              )
  return response["choices"][0]["message"]["content"]
```

```python
prompt = "What is Dina Wang lodgement date?"
prompt_embedding = get_embedding(prompt, engine=embedding_model)
df["similarity"] = df.embedding.apply(lambda x: cosine_similarity(x, prompt_embedding))
results = (df.sort_values("similarity", ascending=False))

results.to_csv('results.csv', index=False)

results
```

```python
messages = [{"role": "system", "content": system},]
messages.append({"role": "user", "content": prepare_prompt(prompt, results)})
len(encoding.encode(''.join(str(message) for message in messages)))

print(messages)

response = answer(messages)
response
```
